## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

```python
# Lane: Refresh / Content Opportunity Scoring

# ML task type: Ranking / Scoring

# The goal is to rank content pages according to how strongly they appear to deserve review.
# The system should produce a priority score for each page so that an SEO or content team can review the highest-priority pages first.

# This is primarily a ranking/scoring problem because the team has limited time and cannot review every page at once.
# The output should therefore help determine which pages should be reviewed first.
```

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

```python
# Target / Proxy:

# For the starter dataset, I will use is_declining_label as a proxy target, where a page is labeled as declining when trend_direction == "down".

# This is only a proxy because the label is derived from the current trend window rather than measuring a future outcome.
# For a stronger future version of the project, I would prefer to define a future-looking target,
# such as whether a page experiences a meaningful decline in the next 30 days based on its previous 90 days of data.

# The final system should ideally rank pages based on the probability or priority of needing attention,
# rather than claiming that a refresh will definitely improve performance.
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

```python
# Success metric: Precision@K, especially Precision@50.

# Precision@50 measures how many of the top 50 pages ranked by the system are actually positive according to the chosen target.

# This metric fits the real business decision because a content team may only have enough capacity to review a limited number of pages.
# If the system recommends 50 pages, we want as many of those 50 recommendations as possible to be relevant.

# I will compare the ML-based ranking against a simple baseline rule to determine whether the model provides additional value.

```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

```python
# A fixed rule can identify obvious cases, such as pages with high impressions and low CTR or pages that have not been updated for a long time.
# However, a page's potential need for review may depend on many signals at the same time, including impressions, CTR, average position, content age, freshness, word count, and engagement.

# ML may be useful because it can combine these signals and learn patterns that are difficult to express with a small number of manually written rules.

# However, ML should only be used if it can demonstrate a meaningful improvement over a transparent baseline.
# I will therefore compare the model's ranking performance with a simple rule-based approach using an appropriate validation method.
```
